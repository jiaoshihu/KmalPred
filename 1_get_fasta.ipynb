```python
import pandas as pd
```

```python
df = pd.read_excel("Train.xlsx", sheet_name="Sheet1")
```

```python
df = df.rename(columns={
    "Uniprot Accession": "Accession",
    "Position": "Site",
    "Sequence": "Sequence"
})
```

```python
df = df.dropna(subset=["Accession", "Sequence", "Site"]).copy()
df["Site"] = df["Site"].astype(int)
```

```python
agg = (
    df.groupby(["Accession", "Sequence"])["Site"]
      .apply(lambda s: sorted(set(int(x) for x in s)))
      .reset_index()
)
```

```python
agg
```

```python
fasta_lines = []
for _, row in agg.iterrows():
    seq = row["Sequence"]
    header = f">{row['Accession']}"       # 大于号后面Accession+位点
    fasta_lines.append(header)
    fasta_lines.append(seq)

# 保存到文件
with open("tain.fasta", "w") as f:
    f.write("\n".join(fasta_lines))
```
